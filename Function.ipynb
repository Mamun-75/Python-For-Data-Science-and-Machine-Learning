```python
## function  defination
def my_func(param1):
  print(param1)
```

```python
my_func('hello')
```

```python
### function practice

def my_func( name ):
  print('hello ' + name);
```

```python
my_func('mamun')   # we can pass value for the parameter directly
```

```python
my_func(name  = 'jose')    # we can mention parameter name + value for more clearity
```

```python
## we can also set default value for parameter
def my_func( name = 'default name' ):
  print('hello ' + name);
```

```python
my_func()
```

```python
my_func('Mamun')
```

```python
### more practice
## we use return statement when the function actually return something
def square(num):
  """
  We can use documentation string to explain function
  Actually this is a multiline string, we can
  use this to explain function in details
  """
  return num**2
```

```python
output = square(2)
```

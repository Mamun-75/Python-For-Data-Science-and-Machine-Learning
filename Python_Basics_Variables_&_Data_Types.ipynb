```python
#slice notation for string
s = 'abcdefghijk'
print( s[0:] )  # print everything starting from index  0
print( s[:3] )  # print everything before index 3
print( s[3:6] ) #print everything starting from index 3
                # and stop printing when find index 6
```

```python
from __future__ import nested_scopes
### list
print([1,2,3])

print(['a','b','c'])

my_list = ['a','b','c']

print(my_list[0] )  # it also supports slice notation

print(my_list[1:3])

# updating list

my_list[0] = "NEW"

print(my_list)

# append
my_list.append('d')
print(my_list)

```

```python
# nested list
nest = [1,2,[3,4]]
print(nest)
print(nest[2])
print(nest[2][1])

nest = [1,2,3,[4,5,['target']]]
print(nest[3][2][0])
```

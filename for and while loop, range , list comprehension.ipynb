```python
seq = [1,2,3,4,5]
```

```python
for item in seq:
  print(item)
```

```python
# whilie loop
i = 1
while i<5:
  print('i is : {}'.format(i))
  i = i+1
```

```python
## list comprehension
```

```python
x  = [1,2,3,4]
```

```python
out = []  # empty list
for num in x:
  out.append(num**2)
```

```python
print(out)
```

```python
# list comprehension is kind of  a for loop but backwards ,
## its easier and short way , it saves some typing for us,
## and create a list in one line !
```

```python
[num**2 for num in x]
```

```python
out = [ num**2 for num in x] # its saying give me the numbers squared for numbers in x
```

```python
## for multiple condition
n = 3

if n == 1:
  print('First')
elif n == 2:
  print('second')
elif n == 3:
  print('third')
else:
  print('last')
```

```python
n = 3

if n == 1:
  print('First')
elif n == 3:
  print('second') ## its true , it will be executed
elif n == 3:
  print('third') ## though its true , but not executed, cz only first true condition is executed
else:
  print('last')
```
